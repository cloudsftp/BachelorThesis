```python
from qiskit.circuit import QuantumCircuit

import matplotlib.pyplot as plt
```

```python
deutsch = QuantumCircuit(2, 1)

deutsch.h([0, 1])
deutsch.cx(0, 1)
deutsch.h(0)

deutsch.measure(0, 0)
```

```python
deutsch.draw(output='latex', filename='../../80_Figures/02_Background/deutsch_algorithm_circuit.png')
```
