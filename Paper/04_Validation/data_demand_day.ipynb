```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
demand_day_df = pd.read_csv('../../Experiments/Data/DemandData/combined_data.csv')
```

```python
print(demand_day_df)
```

```python
fig = plt.figure()
ax = fig.add_axes([0.1, 0.15, 0.8, 0.8])

x = [i for i in range(144)]

ax.plot(x, demand_day_df['power_kW'])

ax.set_xlabel('Time of Day')
ax.set_xticks([24 * i for i in range(7)])
ax.set_xticklabels(['{:2d}:00'.format(4 * i) for i in range(7)])

ax.set_ylabel('Power Demand (kW)')

plt.savefig('../80_Figures/04_Validation/data_demand_day.png')
```
