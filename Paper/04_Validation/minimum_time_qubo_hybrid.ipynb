```python
from dwave.system import LeapHybridSampler
import matplotlib.pyplot as plt
```

```python
sampler = LeapHybridSampler()
time_points = sampler.properties['minimum_time_limit']
```

```python
time_points
```

```python
var = [p[0] for p in time_points]
min_time = [p[1] for p in time_points]
```

```python
plt.plot(var, min_time)
plt.xlabel('Number of Variables')
plt.ylabel('Minimum Computing Time (s)')
plt.savefig('../80_Figures/04_Validation/minimum_computing_time_qubo_hybrid.png')
```

```python
var[-1] = 204800
var = [n / 1024 for n in var]
plt.plot(var, min_time)
plt.xlabel('Number of Loads')
plt.ylabel('Minimum Computing Time (s)')
plt.savefig('../80_Figures/04_Validation/minimum_computing_time_qubo_hybrid_200.png')
```
