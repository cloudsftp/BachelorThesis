```python
import pandas as pd
import matplotlib.pyplot as plt
```

# D-Wave Qbits

```python
qbits = [128, 512, 1152, 2048, 5640]
year = [2011, 2013, 2015, 2017, 2020]

plt.plot(year, qbits)
plt.xlabel('Year')
plt.ylabel('Qubits')
plt.savefig('../../80_Figures/02_Background/dwave_qbits_history.png')
```

# IBM Quantum Volume

```python
qv = [4, 8, 16, 32]
year = [2017, 2018, 2019, 2020]

plt.plot(year, qv)
plt.xlabel('Year')
plt.xticks(year)
plt.ylabel('Quantum Volume')
plt.savefig('../../80_Figures/02_Background/ibm_qv_history.png')
```
